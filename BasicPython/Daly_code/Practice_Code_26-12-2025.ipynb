## 13. Comparing two string for ANAGRAMS

```python
str1 = "listen"
str2 = "silent"

str1 = str1.replace(" "," ").upper()
str2 = str2.replace(" ", " ").upper()

if sorted(str1) == sorted(str2):
    print(True)
else:
    print(False)
```

## 14. Checking for palindrome string

```python
str1 = "Kayak".lower()
if str1 == str1[::-1]:
    print("Palindrome is True")
else:
    print("Palindrome is False")
```
