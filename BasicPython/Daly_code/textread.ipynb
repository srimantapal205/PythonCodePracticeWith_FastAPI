```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *  # Import the function
spark = SparkSession.builder.getOrCreate()
from pyspark.sql.functions import regexp_replace, col
```

```python
filePath =  'C:\\Users\\srima\\Documents\\learning\\Technology\\python\\FastAPI\\PythonCodePractice\\BasicPython\\Daly_code\\file.txt'
fileRead = spark.read.text(filePath)

# Remove dot and space after dot and trim
fileRead = fileRead.withColumn('value', regexp_replace(col('value'), r'[^a-zA-Z0-9\s]', ' '))
fileRead = fileRead.withColumn('value', trim(col('value')))
fileRead.show(truncate=False)


```

```python

df = fileRead.withColumn('words', explode(split(col('value'), ' '))).drop('value')
df.show(truncate=False)
```

```python

df = df.withColumn('length', length(col('words')))
df.show(truncate=False)
```
