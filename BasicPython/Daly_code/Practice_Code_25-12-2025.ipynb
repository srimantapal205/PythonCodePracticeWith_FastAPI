## 1. Convert an Integer into Decimal.

```python
import decimal
int_num  = 43
print(decimal.Decimal(int_num))
print(type(decimal.Decimal(int_num)))
```

## 2. Conver a string of integer into Decimal.

```python
str_num = "123456789"
print(decimal.Decimal(str_num))
print(type(decimal.Decimal(str_num)))
```

## 3. Reversing a string using an Extended Slice Technique.

```python
string = "Python Programming"
reversed_string = string[::-1]
print(reversed_string)
```

## 4. Count vowels in a string. in a Given String.

```python
vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
def count_vowels(inputString):
    count = 0
    for char in inputString:
        if char in vowels:
            count += 1
    return count

inputString = "Hello, welcome to the world of Python programming!"
print(f"Number of vowels in the the given string: {count_vowels(inputString)}")
```

## 5. Counting CONSONANTS in a Given String.

```python
def count_consonants(inputString):
    count = 0
    for chr in inputString:
        if chr not in vowels:
            count += 1
    return count

inputString = "Hello, welcome to the world of Python programming!"
print(f"Number of consonants in the given string: {count_consonants(inputString)}")
```

## 6 Counting the number of occurrences of a character in a string.

```python
def occurance_char(char, inputString):
    count = 0
    for c in inputString:
        if c == char:
            count +=1
    return count

inputString = "Hello, welcome to the world of Python programming!"
char = 'o'

print(f"Number of occurrences of character '{char}' in the given string: {occurance_char(char, inputString)}")
```

## 7. Writing Fibonacci series. 

```python
# Range starts from 0 by default
def fibonacci(n):
    fib = [0, 1]
    for i in range(n):
        next_fib = fib[-1] + fib[-2]
        fib.append(next_fib)
    return fib

print(fibonacci(5))
```

## 8. Finding the maximum value in a list.

```python
def get_maximum(inputList):
    max_value = inputList[0]
    for num in inputList:
        if max_value < num:
            max_value = num
    return max_value

inputList = [3, 5, 2, 8, 1, 4]
print(f"The maximum value in the list is: {get_maximum(inputList)}")
```

## 9. Finding the minimum value in a list.

```python
def get_minimum_value(inputList):
    min_value = inputList[0]
    for num in inputList:
        if min_value > num:
            min_value = num
    return min_value

inputList = [3, 5, 2, 8, 1, 4]
print(f"The minimum value in the list is: {get_minimum_value(inputList)}")
```

## 10. Finding the middle element in list. 

```python
def middle_element(inputList):
    length = len(inputList)
    midElement = inputList[length // 2]
    return midElement

inputList = [3, 5, 2, 8, 1, 4, 7]
print(f"The middle element in the list is: {middle_element(inputList)}")
```
